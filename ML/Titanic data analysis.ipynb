```python
def age_to_category (age) :
    if age < 4 : 
        return 0 
    elif age < 10 : 
        return 1
    elif age < 21: 
        return 2
    elif age < 33 : 
        return 3
    elif age < 50 : 
        return 4
    else : 
        return 5 
    
age_cats = {
        0 : 'babies' , 
        1 : 'children',
        2 : 'teen',
        3 :  'young adult' ,  
        4 : 'adult' , 
        5 : 'elder '
    } 
    
    
    
    
```

```python
# adding new culmn to our  data frame 

df['age_phase'] = df ['age']
```

```python
for i in range (df.shape[0]) : 
    df['age_phase'].iloc[i]= age_to_category(df['age'].iloc[i])
    
```

```python

df['age_phase'].hist(bins = 20 , figsize=(10 , 8))
```

```python
for phase in   df['age_phase'].unique():

    print (age_cats[phase])
    df_phase = df[df['age_phase'] == phase ]
    survived = df_phase[df_phase['survived'] == 1 ]
    survived_percentage = (survived.shape[0] / df_phase.shape[0] ) * 100 
    
    print ( "count : "  , df_phase.shape[0] )
    print ( 'survived : ' ,"%.2f" %  survived_percentage , ' %')
    
    print ('\n===========\n')
```
