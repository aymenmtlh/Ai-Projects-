```python
import pandas as pd 
import numpy as np  
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
df =  pd.read_csv('Telco-Customer-Churn.csv')
```

```python
df.drop('customerID' , axis = 'columns' , inplace = True )
```

```python
df1 = df[df.TotalCharges!=' ']
```

```python
df1.TotalCharges =  pd.to_numeric(df1.TotalCharges)
```

```python
df1.TotalCharges.dtype
```

```python
no_churn_tenure = df1[df1.Churn =='No'].tenure
yes_churn_tenure = df1[df1.Churn =='Yes'].tenure
```

```python
plt.hist([no_churn_tenure , yes_churn_tenure] , color = ('green'  , 'red')  ,  label =['no_churn'  , 'yes_churn']  )
plt.legend()
plt.xlabel('Tenure ')
plt.ylabel('number of customers')
plt.title('churn priduction visualition ')
plt.show()
```

```python
def print_unique_column_val (df) : 
    for column in df :
        if df[column].dtype == 'object' :
            print(f'{column}  : {df[column].unique()}')
```

```python
print_unique_column_val (df1
                        )
```

```python
df1.replace('No phone service'  ,'No' , inplace= True  ) 
df1.replace('No internet service'  ,'No' , inplace= True  ) 
```

```python
print_unique_column_val (df1
                        )
```

```python
yes_no_columns = ['Partner' ,'Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV',
                 
                 'StreamingMovies' ,'PaperlessBilling' ,'Churn' , '']
for  col in yes_no_columns :
    df1[col].replace({'Yes' : 1 , 'No' : 0  } ,  inplace = True)
```

```python

for column in df1 :
    print(f'{column}  : {df1[column].unique()}')
```

```python
df1['gender'].replace({'Female' : 1 , 'Male': 0} , inplace =  True )
```

```python
df2 = pd.get_dummies( data= df1 ,  columns= ['InternetService','Contract' ,'PaymentMethod'])
```

```python
df2.dtypes
```

```python
##scalling 
```

```python
cols_to_scales = ['MonthlyCharges','TotalCharges','tenure']

from sklearn.preprocessing import MinMaxScaler

scaler  =  MinMaxScaler()
df2[cols_to_scales] = scaler.fit_transform(df2[cols_to_scales])
```

```python
X = df2.drop('Churn' , axis = 'columns' )
y = df2['Churn']



```

```python
from sklearn.model_selection import train_test_split 

X_train , X_test , y_train , y_test =  train_test_split(X , y , test_size =0.2 ,  random_state=5)
```

```python
import tensorflow as tf 
from tensorflow import keras 
```

```python
model  =  keras.Sequential([
    
    keras.layers.Dense(20  , input_shape = (26 ,  )  , activation = 'relu'),
    
    keras.layers.Dense(1, activation = 'sigmoid')
    
])
```

```python
model.compile(     
                optimizer='adam' , 
                loss= 'binary_crossentropy', 
                metrics = ['accuracy']
)
```

```python
model.fit(X_train , y_train , epochs=100)
```

```python
model.evaluate(X_test , y_test )
```

```python
yp =  model.predict(X_test)
```

```python
yp[:5]
```

```python
y_pred = []


for val in  yp : 
    if val > 0.5 : 
        y_pred.append(1)
    else :
        y_pred.append(0)
        
        
```

```python
y_pred [:5]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix 
import seaborn  as sn 
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
